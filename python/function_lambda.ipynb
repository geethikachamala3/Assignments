lambda function is a small anonymous function defined using the lambda keyword

if will take any number of arguments
it can have one one expression
it automatically return the result expression
return keyword is not required


```python
def add(a, b):
  return a+b
```

```python
add(3, 4)
```

```python
addition=add(5,6)
```

```python
addition
```

```python
addition1=lambda a, b: a+b
```

```python
addition1(3,6)
```

```python
models=[
    {"name": "RandomForest", "accuracy":0.89},
    {"name": "XGBoost", "accuracy":0.92},
    {"name": "SVM", "accuracy": 0.85}
]
```

```python
dictionary=list(models[0].items())
dictionary
```

```python
dictionary=list(models[1].items())
dictionary
```

A function is a reusable block of code that performs a specific task
instead pof writing the same code multiple times you write it once in a function and call it whenever needed


```python
#calculate avg of three groups
group1_sum=12+34+57
group1_avg=group1_sum/3
print(group1_avg)
group2_sum=45+67+23
group2_avg=group2_sum/3
print(group2_avg)
group3_sum=43+78+89
group3_avg=group3_sum/3
print(group3_avg)
```

```python
#with functions
def calculate_average(scores):
  total=sum(scores)
  avg=total/len(scores)
  return avg
print(f"Grouped 1 average: {calculate_average([80,23,56])}")
```

```python
calculate_average([23,67,89,12])
```

```python
calculate_average([12,13,14,15])
```

```python
#parameters=variables listed in the function definiton
#Arguments=actual values passed when calling the function
#1 positional arguments
#arguments passed in the correct order
def calculate_bmi(weight, height):
  """calculate body mass index"""
  bmi=weight/(height**2)
  return bmi
```

```python
bmi1=calculate_bmi(70,1.75)
print(bmi1)
```

```python
bmi2=calculate_bmi(75, 2.02)
print(bmi2)
```

```python
bmi1+bmi2
```

```python
#keywoed argument
#arguments passed by parameter name (order doesnt match)
def student_info(name, age, grade):
  """display student info"""
  print(f"Name: {name}")
  print(f"Age: {age}")
  print(f"Grade: {grade}")
```

```python
#variable length arguments(*args)
def calculate_sum(*numbers):
  total=0
  for num in numbers:
    total+=num
  return total
```

```python
calculate_sum(1,2,3,4,5,6,7,8,9)
#print(calculate_sum)
```

```python
def print_args(*args):
  print(f"Types: {type(args)}")
  print(f"Values: {args}")
```

```python
print_args(1,2,3,4)
```

```python
#variable length keyword args(**kwords)
#accept any number of keyword arguments
def student_detail(**details):
  """Displays student details"""
  print("Student info")
  for key,value in details.items():
    print(f"{key}: {value}")
```

```python
student_detail(name="Mounisha",age=22, grade="A",city="Chittoor")
```

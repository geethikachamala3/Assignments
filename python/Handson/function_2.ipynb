```python
#combining all arguments types
#order is imp:positional,*args,default,**kwargs
def complex_functions(a,b,*args,c=10,**kwargs):
  print(f"a (position): {a}")
  print(f"b (position): {b}")
  print(f"args (variable position): {args}")
  print(f"c (default): {c}")
  print(f"*kwargs (variable keyword): {kwargs}")
```

```python
complex_functions(1,2,3,4,5,c=20,x=100,y=200)
```

```python
#multiple return values
def calculate_state(numbers):
  minimum=min(numbers)
  maximum=max(numbers)
  average=sum(numbers)/len(numbers)
  return minimum,maximum,average
```

```python
numbers=[34,2,35,67,8,76,90,21,1,23,4,5]
result=calculate_state(numbers)
print(result)
```

```python
min_value,max_value,avg_value=calculate_state(numbers)
```

```python
def calculate_mode(numbers):
    frequency = {}

    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1

    max_freq = max(frequency.values())

    # if max freqeuncy is 1 = No mode
    if max_freq == 1:
        return None

    modes = [num for num, freq in frequency.items()]
    return modes
```

```python
numbers=(1,2,3,5,6,4,3,7,9,8,5,4)
mode=calculate_mode(numbers)
mode
```

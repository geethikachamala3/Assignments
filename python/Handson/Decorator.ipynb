```python
#basic decorator structure
def decorator_function(original_function):
  def wrapper():
    print("Before function runs")
    original_function()
    print("After function runs")
  return wrapper
```

```python
decorated=decorator_function(greet)
```

```python
decorated()
```

```python
#basic sysntax of a decorator
def decorator_name(orginal_function):
  def wrapper(*args, **kwargs):
    ##code befor function runs
    result=original_function(*args, **kwargs)
    ##code after function runc
    return result
  return wrapper
```

```python
@decorator_name
def my_function1():
  print("hello")

@decorator_name
def my_function2():
  print("hello")

@decorator_name
def my_function3():
  print("hello")

@decorator_name
def my_function4():
  print("hello")
```

```python
#another way of decorator function
def my_function1():
  print("hello")
my_function=decorator_name(my_function1)
my_function()
```

```python
#logginf decorator
def log_decorator(func):
  def wrapper(*args, **kwargs):
    print(f"Calling function: {func.__name__}")
    print()
    result=func(*args, **kwargs)
    print()
    print(f"Finished function: {func.__name__}")
    return result
  return wrapper
```

```python
def greet(name):
  print(f"Calling function: {greet.__name__}")
  print(f"Hello, {name}")
  print(f"Finished function: {greet.__name__}")
  print()
```

```python
greet("Mounisha")
```

```python
greet("Manisha")
```

```python
import time
def timer_decorator(func):
  def wrapper(*args, **kwargs):
    start=time.time()
    result=func(*args, **kwargs)
    end=time.time()
    print(f"Time take: {end-start} seconds")
    return result
  return wrapper
```

```python
def greet_name(name, age):
  print(f"Hello {name}, you are {age} year old")
```

```python
greet_name("Joshi", 3)
```

```python
@timer_decorator
def slow_function():
  time.sleep(3)
  print("Done")
slow_function()
```

```python
@timer_decorator
def slowest_function():
  time.sleep(4)
  print("Done execution")
slowest_function()
```

```python
@timer_decorator
def fast_function():
  time.sleep(1)
  print("Executed fact function")
fast_function()
```

```python
print(slow_function())
print(slowest_function())
print(fast_function())
```

```python
def positive_only(func):
  def wrapper(num):
    if num<0:
      return "only positive number allowed"
    return func(num)
  return wrapper
```

```python
@positive_only
def square(num):
  return num*num
```

```python
print(square(2))
```

```python
print(square(0))
```

```python
print(square(-5))
```

```python
def decorator1(func):
  def wrapper(*args, **kwargs):
    print("Decorator 1 - Before")
    result=func(*args, **kwargs)
    print("Decorator 1 - After")
    return result
  return wrapper

def decorator2(func):
  def wrapper(*args, **kwargs):
    print("Decorator 2 - before")
    result=func(*args, **kwargs)
    print("Decoratoe 2 - After")
    return result
  return wrapper

```

```python
@decorator2
@decorator1
def greet():
  print("Hello")
```
